```python
df6['Año'] = df6['Año'].astype('category')
df6['Trimestre'] = df6['Trimestre'].astype('category')
```

```python
df6.info()
```

###El siguiente gráfico de cajas muestra las variaciones en los datos en los distintos rangos de velocidad. Se observa claramente que en los rangos de 1 a 6 Mbps y +30 Mbps es donde hay mayor dispersión de los datos.

```python
numeric_columns = ['HASTA 512 kbps', '+ 512 Kbps - 1 Mbps', '+ 1 Mbps - 6 Mbps', '+ 6 Mbps - 10 Mbps',
                   '+ 10 Mbps - 20 Mbps', '+ 20 Mbps - 30 Mbps', '+ 30 Mbps', 'OTROS']
data_for_boxplot = df6[numeric_columns]

plt.figure(figsize=(8, 4))
sns.boxplot(data=data_for_boxplot, orient='vertical')
plt.xlabel('Categoría de Velocidad')
plt.ylabel('Velocidad de Internet')
plt.title('Distribución de velocidades de Internet')
plt.xticks(rotation=90)
plt.show()
```

###En el siguiente gráfico de barras apiladas muestra cómo se distribuyen las diferentes velocidades de Internet a lo largo de los años. Las velocidades más predominantes son "1 a 6 Mbps" y "+30 Mbps".

```python
numeric_columns = ['HASTA 512 kbps', '+ 512 Kbps - 1 Mbps', '+ 1 Mbps - 6 Mbps', '+ 6 Mbps - 10 Mbps',
                   '+ 10 Mbps - 20 Mbps', '+ 20 Mbps - 30 Mbps', '+ 30 Mbps', 'OTROS']
grouped_data = df6.groupby('Año')[numeric_columns].sum()

plt.figure(figsize=(10, 6))
bottom = None

for column in numeric_columns:
    plt.bar(grouped_data.index, grouped_data[column], bottom=bottom, label=column)
    if bottom is None:
        bottom = grouped_data[column]
    else:
        bottom += grouped_data[column]

plt.xlabel('Año')
plt.ylabel('Velocidad de Internet')
plt.title('Distribución de velocidades de Internet por año')

plt.legend()
plt.show()
```

###Al ser "1 a 6 Mbps" una de los rangos de velocidad más predominantes, se realiza un gráfico de línea donde se muestran la distribución de esta por provincia. Claramente se nota que fue disminuyendo su uso, a diferencia de la que supera los 30 Mbps.

```python
provinces = df6['Provincia'].unique()
data_by_province = df6.groupby(['Provincia', 'Año'])[numeric_columns].sum()

plt.figure(figsize=(10, 6))

for province in provinces:
    province_data = data_by_province.loc[province]
    plt.plot(province_data.index, province_data['+ 1 Mbps - 6 Mbps'], label=province)

plt.xlabel('Año')
plt.ylabel('Velocidad de Internet')
plt.title('Evolución de la velocidad de Internet (+ 1 Mbps - 6 Mbps) por provincia')
plt.legend()
plt.show()
```

###Con el mismo criterio del gráfico anterior se realiza lo mismo para el rango "+30 Mbps" y se observa un claro aumento de esta velocidad con el avance de los años. Esto puede ser causado por diferentes factores, pero entre los mas acertados pueden estar la pandemia, a partir del 2020, los avances tecnológicos y las plataformas de streaming.

```python
provinces = df6['Provincia'].unique()
data_by_province = df6.groupby(['Provincia', 'Año'])[numeric_columns].sum()

plt.figure(figsize=(10, 6))

for province in provinces:
    province_data = data_by_province.loc[province]
    plt.plot(province_data.index, province_data['+ 30 Mbps'], label=province)

plt.xlabel('Año')
plt.ylabel('Velocidad de Internet')
plt.title('Evolución de la velocidad de Internet (+ 30 Mbps) por provincia')
plt.legend()
plt.show()
```

```python
df_ingresos_anuales = df3.groupby('Año')['Ingresos (miles de pesos)'].sum().reset_index()
df_combinado = pd.merge(df_ingresos_anuales, df18, on='Año', how='left')
df_combinado['Ingresos_Ajustados'] = df_combinado['Ingresos (miles de pesos)'] / (df_combinado['Indice de inflación de servicios'].shift() / 100 + 1)
df_combinado['Incremento_Real'] = df_combinado['Ingresos_Ajustados'].pct_change() * 100
```
